### Stability function and and <a href="https://en.wikipedia.org/wiki/Pad%C3%A9_approximant">Padé approximant</a>
<br>

<b>Problem 3.</b> Determine the stability functions of Problem 1.'s methods based on the learned theorem.

```octave
% Butcher tableau
A=[0 0; 1/2 1/2];
b=[1/2 1/2]';
e=ones(size(A,1),1);
c=sum(A,2);

% Order conditions up to order six (naive approach)
p1=[b'*e-1];
p2=[b'*c-1/2];
p3=[b'*c.^2-1/3, b'*A*c-1/6];
p4=[ b'*c.^3-1/4, b'*diag(c)*A*c-1/8, b'*A*c.^2-1/12, b'*A^2*c-1/24];
p5=[c'.^4*b-1/5,
   (b.*c.^2)'*A*c-1/10,
   b'*(A*c).^2-1/20,
   (b.*c)'*A*c.^2-1/15,
   b'*A*c.^3-1/20,
   (b.*c)'*A^2*c-1/30,
   b'*A*diag(c)*A*c-1/40,
   b'*A^2*c.^2-1/60,
   b'*A^3*c-1/120];
p6=[c'.^5*b-1/6,
   b'*diag(c).^3*A*c-1/12,
   b'*diag(c)*(A*c).^2-1/24,
   b'*diag(c).^2*A*c.^2-1/18,
   b'*((A*c.^2).*(A*c))-1/36,
   b'*diag(c)*A*c.^3-1/24,
   b'*A*c.^4-1/30,
   b'*diag(c).^2*A^2*c-1/36,
   b'*((A^2*c).*(A*c))-1/72,
   b'*diag(c)*A*diag(c)*A*c-1/48,
   b'*A*diag(c).^2*A*c-1/60,
   b'*A*(A*c).^2-1/120,
   b'*diag(c)*A^2*c.^2-1/72,
   b'*A*diag(c)*A*c.^2-1/90,
   b'*A^2*c.^3-1/120,
   b'*diag(c)*A^3*c-1/144,
   b'*A*diag(c)*A^2*c-1/180,
   b'*A^2*diag(c)*A*c-1/240,
   b'*A^3*c.^2-1/360,
   b'*A^4*c-1/720];
any(abs(p1)>eps)
any(abs(p2)>eps)
any(abs(p3)>eps)
any(abs(p4)>eps)
any(abs(p5)>eps)
any(abs(p6)>eps)

s = length(b);
e=ones(s,1);
p=poly(A-e*b') %Numerator
q=poly(A) % Denumerator

%Preliminaries (taken from SSPsite.org)
    bounds=[-5 5 -5 5];
    dx=(bounds(2)-bounds(1))/500;
    xa=bounds(1):dx:bounds(2); ya=bounds(3):dx:bounds(4);

    X=repmat(xa,length(ya),1); Y=repmat(ya',1,length(xa));
    XY=(X+1i*Y);
    S=ones(size(XY)); T=ones(size(XY));
    m=length(p)-1; n=length(q)-1;

    % Evaluating the numerator
    XYP=XY;
    for j=1:m
      S=S+p(j+1)*XYP; XYP=XYP.*XY;
    end
    % Evaluating the denumerator
    XYP=XY;
    for j=1:n
      T=T+q(j+1)*XYP; XYP=XYP.*XY;
    end

    % Determine the stability function $R(z)=|S(z)/T(z)|$
    R=abs(S./T);

%Plot
contourf(X,Y,1-R, [0 0], 'LineWidth', 1);
set(gca,'FontSize', 10, 'CLim', [0 1]);
colormap([.1 .5 .3; 0 0 0; 1 1 1]);
hold on;
plot([-5 5], [0 0], '--k', 'LineWidth', 1);
plot([0 0], [-5 5], '--k', 'LineWidth', 1);
axis('square');
title('Stability region')
```
