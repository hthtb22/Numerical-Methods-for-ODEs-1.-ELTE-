<b>Problem 5.</b> <a href="https://en.wikipedia.org/wiki/Lorenz_system">Lorenz system</a> (1963)<br>

The model is a system of three ordinary differential equations now known as Lorenz equations

\begin{cases}
x'(t)=\sigma(y(t)-x(t))&\\
y'(t)=rx(t)-y(t)-x(t)z(t)&\\
z'(t)=x(t)y(t)-bz(t)&
\end{cases}

where $\sigma, r, b \in\mathbb{R}^+$ are the parameters of the system. Let us follow Lorenz's orginal discovery, so $\sigma=10$, $b=8/3$, $r=28$ and $[x(0)\ y(0)\ z(0)]'=[0\ 1\ 0]'$.<br><br> 

Apply the forward Euler and RK4 methods to this system. Plot the $x-z$ plane, so the so-called ''Lorenz attractor".
